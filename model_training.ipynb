### Making pipeline

## Random Forest Classifier

```python
# define columns for ordinal encoding
ordinal_col = ['EDUCATION'] 

# define columns for one hot encoding
cat_col = ['MARITALSTATUS', 'GENDER', 'last_prod_enq2','first_prod_enq2']

#define columns for standardization
num_col=feature.select_dtypes('number').columns

# Define the pipeline for numerical features
num_pipe = Pipeline(steps=[
    ('scale', StandardScaler())
])

# Define the pipeline for categorical features
cat_pipe = Pipeline(steps=[ # (name, transform)
    ('onehot', OneHotEncoder(drop='first', handle_unknown='ignore', sparse_output=False))
])

# Define the pipeline for ordinal features
ordi_pipe = Pipeline(steps=[
    ('ordinal', OrdinalEncoder(categories=[['SSC', '12TH', 'OTHERS', 'GRADUATE', 'UNDER GRADUATE', 'PROFESSIONAL', 'POST-GRADUATE']]))
])

# Combine the transformers in a ColumnTransformer
process = ColumnTransformer(transformers=[ #(name, transformer, columns)
    ('num_tran', num_pipe, num_col),
    ('cat_tran', cat_pipe, cat_col),
    ('ori_tran', ordi_pipe, ordinal_col)
], remainder='passthrough')

# Define the final pipeline with the preprocessing and model
final_rf = Pipeline([
    ('process', process),
    ('model', RandomForestClassifier())
])

# Fit the final pipeline
final_rf.fit(x_train, y_train)
```

```python
x_train.shape
```

```python
y_pred_rf=final_rf.predict(x_test)
y_pred_rf
```

```python
def evaluation(actual,pre):
    acc=accuracy_score(actual,pre)
    pr=precision_score(actual,pre,average='macro')
    re=recall_score(actual,pre,average='macro')
    f_score=f1_score(actual,pre,average='macro')
    c_f=confusion_matrix(actual,pre)
    
    return {
        "accuracy":acc,
        "pre":pr,
        "recal":re,
        "f_score":f_score,
        "matrix":c_f
    }
```

```python
evaluation(actual = y_test,pre= y_pred_rf)
```

```python
evaluation(y_test,y_pred_rf)['matrix']
```

```python
from sklearn.metrics import classification_report
# Generate classification report
report = classification_report(y_test, y_pred_rf)

print(report)
```

```python
accuracy_rf = accuracy_score(y_test, y_pred_rf)
round(accuracy_rf,2)
```

```python
precision, recall, f1_score, _ = precision_recall_fscore_support(y_test, y_pred_rf)

for i, v in enumerate(['p1', 'p2', 'p3', 'p4']):
    print(f"Class {v}:")
    print(f"Precision: {precision[i]}")
    print(f"Recall: {recall[i]}")
    print(f"F1 Score: {f1_score[i]}")
    print()
    
```

### Changing Best Hyperparameter for random Forest classifier

```python
from skopt import BayesSearchCV

# define columns for ordinal encoding
ordinal_col = ['EDUCATION'] 

# define columns for one hot encoding
cat_col = ['MARITALSTATUS', 'GENDER', 'last_prod_enq2','first_prod_enq2']

#define columns for standardization
num_col=feature.select_dtypes('number').columns

# Define the pipeline for numerical features
num_pipe = Pipeline(steps=[
    ('scale', StandardScaler())
])

# Define the pipeline for categorical features
cat_pipe = Pipeline(steps=[ # (name, transform)
    ('onehot', OneHotEncoder(drop='first', handle_unknown='ignore', sparse_output=False))
])

# Define the pipeline for ordinal features
ordi_pipe = Pipeline(steps=[
    ('ordinal', OrdinalEncoder(categories=[['SSC', '12TH', 'OTHERS', 'GRADUATE', 'UNDER GRADUATE', 'PROFESSIONAL', 'POST-GRADUATE']]))
])

# Combine the transformers in a ColumnTransformer
process = ColumnTransformer(transformers=[ #(name, transformer, columns)
    ('num_tran', num_pipe, num_col),
    ('cat_tran', cat_pipe, cat_col),
    ('ori_tran', ordi_pipe, ordinal_col)
], remainder='passthrough')


# Define parameter space
param_space = {
    'n_estimators': (50, 200),  # Number of trees in the forest
    'criterion': ['gini', 'entropy'],  # Function to measure quality of split
    'max_depth': (3, 20),  # Maximum depth of the tree
    'min_samples_split': (2, 10),  # Minimum samples required to split a node
    'min_samples_leaf': (1, 5),  # Minimum samples required at a leaf node
    'max_features': ['sqrt', 'log2'],  # Number of features to consider at each split
    'bootstrap': [True],  # Method to sample data points (with replacement)
    'oob_score': [True, False],  # Use out-of-bag samples to estimate the R^2 on unseen data
}


# Initialize and train BayesSearchCV with RandomForestClassifier and parameter space
model = RandomForestClassifier(random_state=42)
bayes_cv = BayesSearchCV(model, param_space, n_iter=30, cv=5, n_jobs=-1, random_state=42)

# Define the final pipeline with the preprocessing and model
final_rf = Pipeline([
    ('process', process),
    ('model', bayes_cv)
])

# Fit the model
final_rf.fit(x_train, y_train)

best_model = final_rf.best_estimator_

# Predict on test set
y_pred = best_model.predict(x_test)

# Calculate accuracy
test_accuracy = accuracy_score(y_test, y_pred)

print(f"Best hyperparameters: {final_rf.best_params_}")
print(f"Test accuracy: {test_accuracy:.2f}")
```
